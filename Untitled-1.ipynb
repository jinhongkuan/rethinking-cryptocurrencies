### 2.2 On Whales

$
r_x \text{: one-sided liquidity for Token X} \\
r_y \text{: one-sided liquidity for Token Y} \\
G \text{: maximum extractable Token Y} \\ 
\Delta X \text{: inflow of Token X into market circulation} \\ 
x \text{: Token X balance} \\ 
$

```python
def comparative_statics_xliquidity_max_extractable(X, Delta_X, xaxis, ylim=None, r_y=1, name='output.png'):

    fig = plt.figure() 
    fig.patch.set_facecolor('white')
    lstyles = ['-', '--', '-.', ':', '']

    for i, (x, delta_X) in enumerate(zip(X, Delta_X)): 
        formula = lambda x_: \
            x_ * r_y * delta_X / ((x_ + x) * (x_ + x - delta_X)) 
        yaxis = formula(xaxis)
        if ylim:
            plt.ylim(ylim)
        plt.plot(xaxis, yaxis, color='black', ls=lstyles[i], label="$x: {0}, \Delta X: {1}$".format(x, delta_X))
    
    plt.legend()
    plt.xlabel('$r_x$')
    plt.ylabel("$v' - v$")
    plt.savefig(name)
    
Xaxis = np.linspace(5, 20)
comparative_statics_xliquidity_max_extractable([3.0, 6.0], [5.0, 5.0], Xaxis, name='positive_delta_X.png')
```
